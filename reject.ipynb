```python
import csv
import numpy as np
import pandas as pd
```

```python
reject_msg = "Sorry, you did not pass our hidden quality assurance tests."

def reject_ass(expname):
    input_name = 'results/{}-1.csv'.format(expname)
    print("Input", input_name)
    data = pd.read_csv(input_name)
    answer_cols = sorted([c for c in data.columns if c[:6] == 'Answer'])
    test1 = data[answer_cols[0]] > data[answer_cols[2]]
    test2 = data[answer_cols[0]] > data[answer_cols[3]]
    test3 = data[answer_cols[1]] > data[answer_cols[2]]
    test4 = data[answer_cols[1]] > data[answer_cols[3]]
    wrong = data[~test1 | ~test2 | ~test3 | ~test4]
    wrong.loc[:, 'Reject'] = reject_msg
    wrong = wrong.set_index('HITId')
    print("num rejects", wrong.shape[0])
    output_name = "results/{}-1-rejected.csv".format(expname)
    wrong.to_csv(output_name, quoting=csv.QUOTE_ALL)
    print("Output", output_name)
```

```python
test_numbers = ['0094', '0096', '0102', '0153']
exps = ['dw' + n for n in test_numbers]
for exp in exps:
    reject_ass(exp)
```
